
# Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
ds=pd.read_csv("dataset_website.csv")
ds
```

Checking for null values - We find there are no null values

```python
ds.isnull().any()
```

We used the describe funtion to check the values and we see that all the values except index vary from 1 to -1 and hence this shows there are no outliers in the dataset.

We checked the types of all columns and all are in numeric form

Checking the correlation between the columns there are no very high correlations

```python
ds.corr()
```

```python
import seaborn as sns
sns.heatmap(ds.corr())
```

We further dropped the index column as it is not a factor on which our result depends

```python
ds=ds.drop('index', axis = 1)
ds
```

Now, we split the dataset into dependent and independent datasets i.e x for dependent and y for independent

```python
x=ds.iloc[:,0:30].values
y=ds.iloc[:,-1].values
x
```

```python
y
```

# Splitting into testing and training datasets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Model Building

The values in y are either 1 or -1 hence it is a categorical column hence we will use classification algorithms.

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predLR=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
lrAccuracy=accuracy_score(y_test,y_predLR)
lrAccuracy
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB(var_smoothing=1e-09)
gnb.fit(x_train,y_train)
```

```python
y_predNB=gnb.predict(x_test)
y_predNB
```

```python
nbAccuracy=accuracy_score(y_test,y_predNB)
nbAccuracy
```

# KNN (we use grid search with p values 1,2 and n_neighbours value from 1 to 10 to find the best model)

p : Power parameter for the Minkowski metric. When p = 1, this is
    equivalent to using manhattan_distance (l1), and euclidean_distance
    (l2) for p = 2. 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_grid=GridSearchCV(estimator=KNeighborsClassifier(),param_grid={'n_neighbors':np.arange(1,10),'p':[1,2]},cv=10)
knn_grid
```

```python
knn_grid.fit(x_train,y_train)
```

```python
knn_grid.best_params_
```

```python
knn=KNeighborsClassifier(n_neighbors=1,p=1)
knn.fit(x_train,y_train)
```

```python
y_predKnn=knn.predict(x_test)
y_predKnn
```

```python
knnAccuracy=accuracy_score(y_test,y_predKnn)
knnAccuracy
```

# SVM - rbf

```python
from sklearn.svm import SVC
svm1=SVC(kernel='rbf')
svm1.fit(x_train,y_train) 
```

```python
y_predSVM1=svm1.predict(x_test)
from sklearn.metrics import accuracy_score
svm_rbfAccuracy=accuracy_score(y_test,y_predSVM1)
svm_rbfAccuracy
```

# SVM - Sigmoid

```python
svm2=SVC(kernel='sigmoid')
svm2.fit(x_train,y_train) 
```

```python
y_predSVM2=svm2.predict(x_test)
from sklearn.metrics import accuracy_score
svm_sigAccuracy=accuracy_score(y_test,y_predSVM2)
svm_sigAccuracy
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='gini',random_state=0)
dt.fit(x_train,y_train)
```

```python
y_predDT=dt.predict(x_test)
y_predDT
```

```python
dtAccuracy=accuracy_score(y_test,y_predDT)
dtAccuracy
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RandomForestClassifier()
```

```python
rf_grid=GridSearchCV(estimator=RandomForestClassifier(),param_grid={'n_estimators':np.arange(1,20),'criterion':['gini','entropy'],'max_depth':np.arange(2,10)},cv=5)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_grid.best_params_
```

```python
rf=RandomForestClassifier(n_estimators=19,max_depth=9)
rf.fit(x_train,y_train)
```

```python
y_predRF=rf.predict(x_test)
y_predRF
```

```python
rfAccuracy=accuracy_score(y_test,y_predRF)
rfAccuracy
```

# Model Comparison

```python
models = pd.DataFrame({ 
    'Model': [ 'Logistic Regression', 'Naive Bayes','KNN','SVM-rbf','SVM-sigmoid','Decision Tree','Random Forest'],
    'Test Score': [ lrAccuracy,nbAccuracy,knnAccuracy,svm_rbfAccuracy,svm_sigAccuracy,dtAccuracy,rfAccuracy]})
models.sort_values(by='Test Score', ascending=False)
```
